```python
import ujson as json
import pandas as pd
import numpy as np
```

```python
records = map(json.loads,open('prefecture.ndjson'))
df = pd.DataFrame.from_records(records)
df
```

```python
df_ai = df[df['prefecture']=='23']
pd.set_option('display.max_rows',400)
print(len(df_ai))
print('----------------')
print(df_ai)
```

```python
date=df_ai['date'].values[0:]
unique_date=np.unique(date)
unique_date
```

```python
df_ai=df_ai.drop('prefecture',axis=1)
df_ai
```

```python
df_ai.to_csv('aichi_vaccine.csv')
```

```python
df_ai_date=df_ai.set_index('date')
df_ai_date
```

```python
aichi_vaccine_total=[]
for day in unique_date:
    vaccine_summary=df_ai_date.query('date=='+'"'+str(day)+'"')['count'].sum()
    print(str(day)+':'+str(vaccine_summary))
    aichi_vaccine_total.append([[day],[vaccine_summary]])
```
