```python
import ujson
import pandas as pd
import numpy as np
import json
import datetime
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
folderName="20210612"
dir_path = "./"+folderName
if not os.path.exists(dir_path):
    os.makedirs(dir_path)
```

```python
def aichi_ndjson(file,date):
    records = map(ujson.loads,open(file))
    df = pd.DataFrame.from_records(records)
    df_ai=df[df['prefecture']=='23']
    df_ai.to_csv(dir_path+'/aichi_vaccine_'+date+'.csv')
```

```python
date = df['date'].values[0:]
unique_date=np.unique(date)
unique_date
```

```python
#df[df['date']==unique_date[40]]
vaccine_per_day=[]
i=0
for day in unique_date:
    if i==0:
        target=df[df['date']==day]
        vaccine_1st=target[target['status']==1]['count'].sum()
        vaccine_2nd=target[target['status']==2]['count'].sum()
        vaccine_total=target['count'].sum()
        #vaccine_agg=vaccine_total
        vaccine_per_day.append([day,vaccine_1st,vaccine_2nd,vaccine_total])
        i+=1
    else:
        target=df[df['date']==day]
        vaccine_1st=target[target['status']==1]['count'].sum()
        vaccine_2nd=target[target['status']==2]['count'].sum()
        vaccine_total=target['count'].sum()
        #vaccine_agg=vaccine_total+vaccine_per_day[i-1][4]
        vaccine_per_day.append([day,vaccine_1st,vaccine_2nd,vaccine_total])
        i+=1
vaccine_per_day
```

```python
df_day=pd.DataFrame(vaccine_per_day,columns=['date','1st','2nd','total'] )
df_day.dtypes
```

```python
df_day['date']=pd.to_datetime(df_day['date'],format='%Y-%m-%d')
df_day.dtypes
```
